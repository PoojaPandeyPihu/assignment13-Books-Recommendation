```python
# converting long data into wide data using pivot table
book_recomm = pd.pivot_table(data = books_data2 , values='Book.Rating',index='Book.Title',columns='User.ID').fillna(0)
book_recomm
```

```python
book_recomm.columns = books_data2['User.ID'].unique()
```

```python
book_recomm
```

```python
book_recomm.corr().round(2)
```

```python
book_recomm1=book_recomm.T
book_recomm1
```

```python
book_recomm1.values
```

# Using Cosine Similarity for Recommedation Engine

```python
user_to_user_cosine = 1 - pairwise_distances(X = book_recomm1.values, metric='cosine')
user_to_user_cosine
```

```python
user_to_user_cosine_df = pd.DataFrame(data=user_to_user_cosine,columns =books_data2['User.ID'].unique(), index = books_data2['User.ID'].unique())
user_to_user_cosine_df
```
